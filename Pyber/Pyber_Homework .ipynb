```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
combined_data = pd.merge(city_data,ride_data,on="city")

# Display the data table for preview
combined_data_city = combined_data.groupby("city")
combined_data_city.head()

```

```python
help(plt.legend)
```

```python
avg_fares = combined_data_city['fare'].mean()
total_rides = combined_data_city['ride_id'].count()
no_drivers = combined_data_city['driver_count'].mean()
city_types = city_data.set_index('city')['type']
cleaned_data = pd.DataFrame({"Total Rides": total_rides,
                             "City Type": city_types,
                             "Average Fare":avg_fares,
                             "No. of Drivers": no_drivers})

urban = cleaned_data.loc[cleaned_data["City Type"]=="Urban"]
suburban = cleaned_data.loc[cleaned_data["City Type"]=="Suburban"]
rural = cleaned_data.loc[cleaned_data["City Type"]=="Rural"]


plt.scatter(urban['Total Rides'],urban['Average Fare'], marker="o", facecolors="coral", edgecolors="black",s=urban['No. of Drivers']*15, alpha=0.8,label='Urban')
plt.scatter(suburban['Total Rides'],suburban['Average Fare'], marker="o", facecolors="lightskyblue", edgecolors="black",
            s=suburban['No. of Drivers']*15, alpha=0.8,label='Suburban')
plt.scatter(rural['Total Rides'],rural['Average Fare'], marker="o", facecolors="yellow", edgecolors="black",
            s=rural['No. of Drivers']*15, alpha=0.8,label='Rural')
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.text(42,36.5,"Note:") 
plt.text(42, 35,"Circle size correlates with driver count per city.")

legend = plt.legend(loc=1,fontsize='small',title = "City Types")
for handle in legend.legendHandles:
    handle.set_sizes([40])

plt.grid()
plt.savefig('BubblePlot.png')
plt.show()
```

```python
type_fares = combined_data.groupby('type')['fare'].sum()
colors = ["gold","lightskyblue","lightcoral"]
labels = ["Rural","Suburban","Urban"]
explode = (0,0,0.1)
plt.pie(type_fares,explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True,startangle=140)
plt.title("% of Total Fares by City Type")
plt.savefig('TotalFaresbyCityType.png')
plt.show()
```

```python
type_fares = combined_data.groupby('type')['ride_id'].count()
colors = ["gold","lightskyblue","lightcoral"]
labels = ["Rural","Suburban","Urban"]
explode = (0,0,0.1)
plt.pie(type_fares,explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True,startangle=140)
plt.title("% of Total Rides by City Type")
plt.savefig('TotalRidesbyCityType.png')
plt.show()
```

```python
type_fares = city_data.groupby('type')['driver_count'].sum()
colors = ["gold","lightskyblue","lightcoral"]
labels = ["Rural","Suburban","Urban"]
explode = (0,0,0.1)
plt.pie(type_fares,explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True,startangle=140)
plt.title("% of Total Drivers by City Type")
plt.savefig('TotalDriversbyCityType.png')
plt.show()
```
